```python
plt.figure(figsize=(20,15))

#subplot of price (position 1: 1st row, 1st column)
plt.subplot(3,2,1)
sns.histplot(data=df, x='price', kde=True, hue='fuelType', color=set_color[0])

#subplot of mileage (position 3: 2nd row, 1st column)
plt.subplot(3,2,3)
sns.histplot(data=df, x='mileage', kde=True, hue='fuelType', color=set_color[2])

plt.tight_layout()
plt.show()
```

#### Findings:<br>
-Distribution of the price is Unimodal and Skewed Distribution.<br>
-Distribution of the mileage is Unimodal and Skewed Distribution.

### ECDF subplots: Empirical Cumulative Distribution Function subplots of all the numeric columns in the dataframe

```python
plt.figure(figsize=(20,15))

#subplot of price (position 1: 1st row, 1st column)
plt.subplot(3,2,1)
sns.histplot(data=df, x='price', cumulative=True, color=set_color[0])
plt.grid()

#subplot of year (position 2: 1st row, 2nd column)
plt.subplot(3,2,2)
sns.histplot(data=df, x='year', cumulative=True, color=set_color[1])
plt.grid()

#subplot of mileage (position 3: 2nd row, 1st column)
plt.subplot(3,2,3)
sns.histplot(data=df, x='mileage', cumulative=True, color=set_color[2])
plt.grid()

#subplot of miles per gallon (position 3: 2nd row, 1st column)
plt.subplot(3,2,4)
sns.histplot(data=df, x='mpg', cumulative=True, color=set_color[3])
plt.grid()

#subplot of engine size (position 4: 2nd row, 2nd column)
plt.subplot(3,2,5)
sns.histplot(data=df, x='engineSize', cumulative=True, color=set_color[4])
plt.grid()

plt.tight_layout()
plt.show()
```

Figure 6 represents the empirical cumulative distribution of price of used Ford cars. The graphs shows us the probability of occurence of price at a given point. For example, the probability of the price being less than $10,000 is approximately 0.4.
